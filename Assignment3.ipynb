◾요구사항 4. 다음 3개의 사이트에 대한 결과를 출력하시오.◦http://www.naver.com
◦http://www.daum.net

```python
import urllib2
import string

source = urllib2.urlopen("http://www.naver.com").read()

string1 = ''
string2 = ''
string3 = ''
string4 = ''

l = []
r = []
k = []
p = []

for line in source.splitlines():
    line = line.lstrip()
    for word in line:
        if word == '<':
            string1 += '\n'
        string1 += word
        if word == '>':
            string1 += '\n'
        
for line in string1.splitlines():
    line = line.lstrip()
    for word in line:
        string2 += word
    l.append(string2)
    string2 = ''

for x in range(0,len(l)):
    if l[x] != '':
        r.append(l[x])

for x in range(0,len(r)):
    if (r[x][0] == '<') or (r[x][0] == '/') or (r[x][0] == '-'):
        r[x] = ''
        
for x in range(0,len(r)):
    if r[x] != '':
        k.append(r[x])
        
for x in range(0,len(k)):
    string3 += k[x]
    string3 += ' '

    
string4 = unicode(string3, encoding='utf-8')

limit = len(string4)
x = 0

while x<limit:
    for y in range(0,len(string.punctuation)):
        if string4[x] == string.punctuation[y]:
            string4 = string4.replace(string4[x],'')
            limit = len(string4)
    x += 1
    
p = string4.split()
words = {}


for i in p:
    if not i in words.keys():
        words[i]=1
    else:
        words[i]=words[i]+1
        
format = "%s : %d"
  
for i in words:
    print format % (i, words[i])
```
